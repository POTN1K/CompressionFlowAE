# Physical Properties
In this jupyter notebook the calculations of some physical properties of the flow is demonstrated.
These properties are plotted accompanied with the velocity feild. You can change the image to display by changing the image varible

```python
from FlowCompression import AE
u_train, u_val, u_test = AE.preprocess()
image = u_test[111]
```

First you can see the velocity field in two different ways

```python
AE.plot_velocity(image)
AE.u_v_plot(image)
```

Secondly, you can display the energy of the image. In lack of all the neccesary information, the energy is calculated by taking the square of the velocity at each point and taking the half of that

```python
AE.plot_energy(image)
```

Lastly, the vorticity in the image is displayed by calculating the curl. The curl computes the cross product of nabla and the velicity. For the derivatives backward, central and forward difference schemes were used.

```python
AE.plot_vorticity(image)
```
